```python
import pandas as pd 
import os 
import json 
import numpy as np
```

### Carregando o dataset de originações de pagamentos

```python
path_base_originations = "originations/originations/"
```

```python
list_filenames = os.listdir(path_base_originations)
```

```python
df_originations = pd.DataFrame() 

for files in list_filenames:
    path_filename = path_base_originations + files 
    
    with open(path_filename, 'r') as f:
        data = f.readlines()
        
        if len(data) != 0:
            for i in json.loads(data[0])["installments"]:
                
                i["originationId"] = json.loads(data[0])["originationId"]
                i["clientId"] = json.loads(data[0])["clientId"]
                i["registerDate"] = json.loads(data[0])["registerDate"]
                
                
                df_originations = df_originations.append(i, ignore_index=True)
                
```

### Seta a variável payment para data de hoje caso seja NaT

```python
today =  lambda x: pd.to_datetime('today').normalize() if pd.isnull(x) else x
```

```python
df_merge_client_payment.paymentDate = df_merge_client_payment.paymentDate.apply(today)
```

### Cria variável dias de atraso

```python
# Caso número de dias seja negativo, então indica que há D dias de atrasos, nese caso, convertemos a variável para positivo
# Caso não haja atraso, então a variável vai ser um número positivo, então convertemos para 0, pois nosso objetivo é encontrar
# atrasos
lags = lambda x: 0 if x > 0 else x * (-1)
```

```python
df_merge_client_payment["lagDays"] = (df_merge_client_payment.dueDate - df_merge_client_payment.paymentDate).dt.days.apply(lags)
```

### Funçao que verifica se o client é um bom pagador dentro do range da coorte

```python
def coorte_customer_is_good_payer(dt_inicio_coorte, dt_fim_coorte, client_df):
    """
        dt_inicio_coorte: data de inicio da analise do cliente
        dt_fim_coorte: data de fim da analise do cliente
        
        client_df tem que conter os campos: dueDate, lagDays
    
    """
    date_threshold_min = pd.to_datetime(dt_inicio_coorte)
    date_threshold_max = pd.to_datetime(dt_fim_coorte)


    for index, row in client_df[client_df.dueDate >= date_threshold_min].iterrows():

        dueDate = row["dueDate"]
        lagDays = row["lagDays"]
        
        
        
        if dueDate > date_threshold_max:
            
            #print(row["clientId"], "Entrou aqui")

            break 

        # Retorna 1 se tiver pelo menos algum atraso no range especificado acima
        # Caso contrario, retorna 0 e o cliente é um bom pagador dentro da coorte
        
        # Mais de 30 dias de atraso
        #print(lagDays, 30)
        if lagDays >= 30:
            #print(row["clientId"], dueDate, date_threshold_max)
            return 1
        
    return 0 
```

### Analisa a coorte dos clientes nos ranges definidos

```python
clients = df_merge_client_payment.clientId.unique()
```

### 1 mês de coorte

```python
dt_inicio_coorte = "2021-01-01"
dt_fim_coorte = "2021-02-01"

dict_1_month_client = {}

for client in clients:
    
    client_df = df_merge_client_payment[df_merge_client_payment.clientId == client].sort_values("dueDate")
    
    isClientWithLags = coorte_customer_is_good_payer(dt_inicio_coorte, dt_fim_coorte, client_df)
    
    if isClientWithLags == 1:
        dict_1_month_client[client] = isClientWithLags
    
    
```

### 3 meses de coorte

```python
dt_inicio_coorte = "2021-01-01"
dt_fim_coorte = "2021-03-01"

dict_3_month_client = {}

for client in clients:
    
    client_df = df_merge_client_payment[df_merge_client_payment.clientId == client].sort_values("dueDate")
    
    isClientWithLags = coorte_customer_is_good_payer(dt_inicio_coorte, dt_fim_coorte, client_df)
    
    if isClientWithLags == 1:
        dict_3_month_client[client] = isClientWithLags
    
    
```
